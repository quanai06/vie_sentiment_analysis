```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
train_df_1=pd.read_csv("../data/train_backtranslate.csv")
train_df_2=pd.read_csv("../data/train_resample.csv")
test_df=pd.read_excel("../data/test_nor_811.xlsx")
valid_df=pd.read_excel("../data/valid_nor_811.xlsx")
```

```python
train_df_2.tail(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

DATA BACK_TRANSLATE WITH API GG TRAINSLATE

```python
full_train=pd.concat([train_df_1,valid_df])
```

```python
#TF-IDF vectorizer
vectorriver=TfidfVectorizer()
X_train=vectorriver.fit_transform(full_train["Sentence"])
y_train=full_train["Emotion"]
X_test=vectorriver.transform(test_df["Sentence"])
y_test=test_df["Emotion"]
```

```python
model=LinearSVC()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

DATA RESAMPLE

```python
train=pd.concat([train_df_2,valid_df])
vectorriver=TfidfVectorizer()
X_train_2=vectorriver.fit_transform(train["Sentence"])
y_train_2=train["Emotion"]
X_test_2=vectorriver.transform(test_df["Sentence"])
y_test_2=test_df["Emotion"]
```

```python
model2=LinearSVC()
model2.fit(X_train_2,y_train_2)
```

```python
y_pred_2=model2.predict(X_test_2)
print (classification_report(y_test,y_pred_2))
```
